# What are we doing ?

In the case of creating **neural networks**, the sample we would use is one where it learns the relationship between two numbers. So, for example, if we are writing code for a function like this, we already know the 'rules'

float hw_function(float x)

{

    float y = (2 * x) - 1;
    
    return y;
    
}

So how do we train a neural network to do the equivalent task? Using data! By feeding it with a set of Xs, and a set of Ys, it should be able to figure out the relationship between them.

This is obviously a very different paradigm than what you might be used to, so let's step through it piece by piece.

# Importing

- importing _tensorflow_ (it bundles together ML & DL models and algorithms).


- importing _keras_ (keras is a high-level Python neural networks library that runs on top of TensorFlow) OR (framework for defining a neural network as a set of Sequential layers).


- importing _numpy_ (helps represent data as array/lists , therefore easy to work).

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

# Creating our Neural Network

- We will create our _Neural Network_ having :
         
         1 layer  (in the NN)
         
         1 neuron (in that layer)
         
         1 value (to input shape)

```python
model = tf.keras.Sequential([keras.layers.Dense(units = 1 , input_shape = [1])])
```

# Compiling the Neural Network

- We have to specify 2 functions :
          
          loss
          
          optimizer
          

- Relationship between numbers in our function is :
          
          y = 2x-1
          

- When the computer is trying to 'learn' the above , it makes a guess...maybe y = 10x + 10 . The _LOSS_ function measures the guessed answers against the known correct answers and measures how well or how badly it did.


- It then uses the _OPTIMIZER_ function to make another guess. Based on how the loss function went, it will try to minimize the loss. At that point maybe it will come up with somehting like y=5x+5, which, while still pretty bad, is closer to the correct result (i.e. the loss is lower).


- We can repeat this by defining the number of _EPOCHS_.


- We will use :

       'MEAN SQUARED ERROR' for the loss
       
       'STOCHASTIC GRADIENT DESCENT' for the optimizer (sgd)

```python
model.compile(optimizer = 'sgd' , loss = 'mean_squared_error')
```

# How do we feed the data ?

- Let's take a few Xs and a few Ys.

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0] , dtype = float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0] , dtype = float)
```

# Training the Neural Network

- Training of the Neural Network means , it **learns**  the relationship between the Xs and Ys is in the model.fit call (in our eg).


- Process of traing : 
           
           makes a guess , measuring how good/bad it is (LOSS).
           
           makes another guess , using the (OPTIMZER).
           
           how many repeatitions ? (no. of EPOCHS).

```python
model.fit(xs , ys , epochs = 500)
```

- ***Now we have a model that has been trained to learn the relationship between X and Y.***

# Making predictions now

```python
model.predict([10.0])
```

- We expected 19 but we got a little under , reason behind this is that we have trained our model with mimimal amount of data (6 in total) , also NN work on the probabilities not exact results.
